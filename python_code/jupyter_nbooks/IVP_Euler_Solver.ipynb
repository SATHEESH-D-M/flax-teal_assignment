# IVP using Euler Method

- procedural programing approach to solve the given IVP naievly.

- This exercise is to get a reference for developing a generalized solver.

```python
"""imports"""
import numpy as np
import matplotlib.pyplot as plt

"""Parameters for the simulation."""
n = 10 # number of steps
domain_start = 0.0 # start of the domain
domain_end = 5.0 # end of the domain
mesh_1d = np.linspace(domain_start, domain_end, n + 1) # mesh points
h = (domain_end - domain_start) / n # step size


# Initial condition
t_0 = 0.0 # initial value of independent variable
y_0 = 1.0 # initial value of dependent variable
def f(t: float, y: float)-> float:
    """Function defining the ODE dy/dt = f(t, y(t))."""
    return np.cos(t) - y
```

```python
"""Euler's method for solving the initial value problem (IVP)."""
def euler_solver(f, t_0: float, y_0: float, h: float, n: int) -> np.ndarray:
    """
    Solve the IVP using Euler's method.

    Parameters:
    f (callable): Function defining the ODE dy/dt = f(t, y(t)).
    t_0 (float): Initial value of independent variable.
    y_0 (float): Initial value of dependent variable.
    h (float): Step size.
    n (int): Number of steps.

    Returns:
    np.ndarray: Array of computed values at each step.
    """
    y = np.zeros(n + 1) # Initialize array to store computed values
    y[0] = y_0 # Set initial condition
    for k in range(0, n):
        t = t_0 + (k) * h
        y[k + 1] = y[k] + h * f(t, y[k])
    return y
```

```python
# Solve the IVP using Euler's method
y = euler_solver(f, t_0, y_0, h, n)
```

```python
# Analytical solution for comparison
def analytical_solution(t: np.ndarray) -> np.ndarray:
    """Analytical solution of the ODE."""
    return np.exp(-t)/2 + (np.sin(t) + np.cos(t))/2
```

```python
y_analytical = analytical_solution(np.linspace(domain_start, domain_end, 10000))
plt.plot(mesh_1d, y, label='Euler Solution',color="red", markersize=3)
plt.plot(np.linspace(domain_start, domain_end, 10000), y_analytical, label='Analytical Solution', color="black", linestyle='--')

plt.legend()
```

```python
print("Euler's method solution:")
for t, y_val in zip(mesh_1d, y):
    print(f"t = {t:.2f}, y = {y_val:.4f}")
```
